```python
import pandas as pd
```

```python
data = "cleaned_data.csv"
# Read the data into a DataFrame
df = pd.read_csv(data)
df = df[df['country'].str.contains('United States', na=False)]
# Display the filtered DataFrame
print(df)
```

```python
b1 = df[df['comp_prim_v2_m'] != 0].groupby('Wealth')['comp_prim_v2_m'].mean().reset_index()
b2 = df[df['comp_lowsec_v2_m'] != 0].groupby('Wealth')['comp_lowsec_v2_m'].mean().reset_index()
b3 = df[df['comp_upsec_v2_m'] != 0].groupby('Wealth')['comp_upsec_v2_m'].mean().reset_index()
b4 = df[df['comp_prim_1524_m'] != 0].groupby('Wealth')['comp_prim_1524_m'].mean().reset_index()
b1
```

```python
merged_df = b1.merge(b2, on='Wealth').merge(b3, on='Wealth').merge(b4, on='Wealth')
merged_df
```

```python
merged_df = merged_df.rename(columns={'comp_prim_v2_m': 'Block1', 'comp_lowsec_v2_m': 'Block2','comp_upsec_v2_m': 'Block3','comp_prim_1524_m': 'Block4'})
merged_df
```

```python
merged_df.to_json("quintile.json", orient='records')
```
